## Q1. What is the relationship between polynomial functions and kernel functions in machine learning algorithms?

### Ans:
### Polynomial functions and kernel functions are related in machine learning algorithms, specifically in the context of kernel methods such as Support Vector Machines (SVMs).
### Polynomial functions can be used as kernel functions in SVMs to implicitly transform the input data into a higher-dimensional feature space. The kernel trick, which is a technique used in SVMs, allows us to compute the dot products between the transformed feature vectors without explicitly calculating the transformed feature vectors. This computational shortcut enables SVMs to efficiently work with high-dimensional feature spaces and handle complex patterns in the data.

## Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?

### Ans:
### Increasing the value of epsilon tends to increase the number of support vectors in SVR. This is because a larger epsilon allows more data points to fall within the margin of tolerance, resulting in a broader region where deviations from the predicted function are acceptable.

### Increasing the value of epsilon in SVR has an impact on the number of support vectors. Support vectors are the data points that fall on or within the margin boundary, which is defined by the support vectors themselves. They are the critical points that determine the regression line or hyperplane.

## Q4. How does the choice of kernel function, C parameter, epsilon parameter, and gamma parameter affect the performance of Support Vector Regression (SVR)? Can you explain how each parameter works and provide examples of when you might want to increase or decrease its value?

### Ans:
## Kernel Function:
### The kernel function is used to map the input data into a higher-dimensional feature space. It allows SVR to handle nonlinear relationships between the input variables and the target variable. Different types of kernel functions include linear, polynomial, radial basis function (RBF), sigmoid, etc.

### For example:
### Linear Kernel: Suitable when the data has a linear relationship.
### RBF Kernel: Appropriate when dealing with complex, nonlinear patterns.

## C Parameter:
### The C parameter controls the trade-off between achieving a small error on the training data and having a larger margin. Conversely, lower values of C allow for a larger margin and tolerate more errors. 
### For example:

### Increase C when you want to prioritize fitting the training data more accurately, especially if you suspect that the data may have outliers.
### Decrease C when you want to allow more errors and emphasize a larger margin, promoting a more generalizable model.

## Epsilon Parameter:

### The epsilon parameter determines the width of the epsilon-insensitive tube. Increasing epsilon expands the tolerance for errors, allowing more data points to fall within the tube. Decreasing epsilon makes the tube narrower, enforcing stricter error tolerance.

### Increase epsilon when you want to allow larger deviations from the regression line, focusing on capturing the overall trend rather than individual data points.

## Gamma Parameter:
### Higher values of gamma make the decision boundary more localized and can result in more complex and intricate decision boundaries. Lower values of gamma make the decision boundary smoother and can lead to a simpler model.

### Increase gamma when you suspect that the target variable's relationship with the input variables is complex and varies over shorter distances.

### Decrease gamma when you want to generalize the model and smooth out the decision boundary, especially if the relationship between the variables is more global.

```python
import pandas as pd
from sklearn.svm import SVC
```

```python
# creating a synthatic dataset
from sklearn.datasets import make_classification

X, y = make_classification(n_classes=2, n_features=2, n_clusters_per_class=2, n_samples=1000, n_redundant=0)
```

```python
import seaborn as sns
sns.scatterplot(x = pd.DataFrame(X)[0], y = pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
## prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
#define parameter range
param_grid = {'C':[0.1, 1, 10, 100, 1000],
             'gamma':[1,0.1,0.01,0.001,0.0001],
             'kernel':['linear']}
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## prediction
y_pred4 = grid.predict(X_test)
print(classification_report(y_test, y_pred4))
print(confusion_matrix(y_test, y_pred4))
print(accuracy_score(y_test, y_pred4))
```
